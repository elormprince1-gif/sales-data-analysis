```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("sales_data.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
df.info()
```

```python
df['age'].fillna(df['age'].mean(), inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df["Revenue"] =  df["quantity"] * df["price"]
df.head()
```

```python
df["Revenue"].sum()
```

```python
df.groupby("city")["Revenue"].sum().sort_values(ascending=False)
```

```python
df.groupby("product")["Revenue"].sum().sort_values(ascending=False)
```

```python
df.groupby("category")["Revenue"].sum().sort_values(ascending=False)
```

```python
df.groupby("gender")["Revenue"].sum().sort_values(ascending=False)
```

```python
df["age"].mean()
```

```python
df["city"].value_counts()
```

```python
daily_sales = df.groupby('order_date')['Revenue'].sum()

daily_sales.plot()
plt.title("Daily Sales Trend")
plt.show()
```

```python
df.groupby('city')['Revenue'].sum().plot(kind='bar')
plt.title("Revenue by City")
plt.show()
```
